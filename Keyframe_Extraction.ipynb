## Time-based keyframe extraction

```python
import cv2, os
```

```python
def getFrame(sec):
    vidcap2.set(cv2.CAP_PROP_POS_MSEC, sec*1000)
    hasFrames,image = vidcap2.read()
    if hasFrames:
      cv2.imwrite(pathframes+"/frame%d.png"  % count , image)
    return hasFrames

Videospath="VideosPath" #VideosPath
TimeBasedImages=[]

for i in os.listdir(Videospath):
  TimeBasedImages.append(i.split(".avi")[0])
for f in os.listdir(Videospath):
  foldername=f.split(".")[0]

  vidcap2 = cv2.VideoCapture(Videospath+'/'+f)
  if foldername in TimeBasedImages:
    pathframes = "Time-basedFramesPath/"+foldername  #Time-basedFramesPath
    os.mkdir(pathframes)

    sec = 0
    frameRate = .25 #//it will capture image in each quarter of a second
    count=1
    success = getFrame(sec)
    while success:
            sec = sec + frameRate
            sec = round(sec, 2)
            success = getFrame(sec)
            count = count + 1
```

## Similarity-based keyframe extraction

```python
from skimage.metrics import structural_similarity as compare_ssim
```

```python
#Source code of calculation of SSIM Similarity:
#https://pyimagesearch.com/2017/06/19/image-difference-with-opencv-and-python/

SimilarityBasedImages="Sim-basedFramesPath/" #Sim-basedFramesPath
TimebasedImages="Time-basedFramesPath/"#Time-basedFramesPath

def getSimilarFrame(TimebasedImages,SimilarityBasedImages):

for folder in os.listdir(TimebasedImages):

    array_size=0
    image_array = [] #original Images
    gray_array = [] #Gray Images

    for myFile1 in os.listdir(TimebasedImages+folder+'/'):
        array_size=array_size+1
        imageA = cv2.imread(TimebasedImages+folder+'/'+myFile1)
        image_array.append(imageA)

    for img in image_array:
        grayimg = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        gray_array.append(grayimg)
    os.makedirs(SimilarityBasedImages+folder)
    for k in range(array_size):
      for i in range(0, len(gray_array), 2):
          j=i+1
          if j==len(gray_array) or j> len(gray_array):
            break
          else:
            (score, diff) = compare_ssim(gray_array[i], gray_array[j], full=True)
            if score > 0.95 :
        
              gray_array.pop(j)
              image_array.pop(j)

    count=1
    for u in range(len(image_array)):

         cv2.imwrite(SimilarityBasedImages+folder+"/frame%d.png" % count ,image_array[u])
         count=count+1

         
getSimilarFrame(TimebasedImages,SimilarityBasedImages)
```
